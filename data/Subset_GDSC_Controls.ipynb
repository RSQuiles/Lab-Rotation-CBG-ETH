```python
import scanpy as sc
import pandas as pd
import numpy as np
import os
```

```python
gdsc_path = "/cluster/scratch/rquiles/GDSC_counts.h5ad"
adata = sc.read_h5ad(gdsc_path)
```

```python
# Filter only the controls and subset to one third
idx = adata.obs[adata.obs["Agg_Treatment"] == "DMSO_TF"].index
sampled_idx = np.random.choice(idx, size=len(idx)//3, replace=False)

adata = adata[sampled_idx].copy()
adata.write("/cluster/scratch/rquiles/GDSC_controls_subset.h5ad")
```
