```python
import numpy as np
import pandas as pd
import os
import pickle 
import sys
```

```python
#Print the current working directory
print("Current working directory: ", os.getcwd())
datapath = '../src/resources/KEGG/'
circuit_backwards_file = datapath + 'genes_per_circuit_backward.csv'

# Load files csv
circuit_backwards = pd.read_csv(circuit_backwards_file, sep=',')
```

```python
circuit_backwards = circuit_backwards.rename(columns={'Unnamed: 0': 'circuit_name'})
circuit_backwards.set_index('circuit_name', inplace=True)
```

```python
circuit_backwards.head()
```

```python
circuit_description = pd.DataFrame(columns=['circuit_name', 'description'])
circuit_description['circuit_name'] = circuit_backwards.index
#count number of circuits and print
print("Number of circuits: ", len(circuit_backwards.index))
```

```python
# Read csv file to store circuit descriptions
circuit_description = pd.read_csv(datapath + 'circuit_description.csv', sep=',')
for circuit_name in circuit_backwards.index:
    # Check if the circuit already has a description (all circuit names are already in circuit_description, but for some the circuit description is empty string or NaN)
    desc_value = circuit_description.loc[circuit_description['circuit_name'] == circuit_name, 'description'].values[0]
    print(f"Checking circuit: {circuit_name}, current description: {desc_value}")
    if pd.isna(desc_value) or desc_value == '':
        # Get user input for circuit description
        description = input(f"Enter a description for {circuit_name}: ")
        circuit_description.loc[circuit_description['circuit_name'] == circuit_name, 'description'] = description
        # Save the circuit description to a CSV file
        circuit_description.to_csv(datapath + 'circuit_description.csv', index=False)
    else:
        print(f"Circuit {circuit_name} already has a description: {desc_value}")
```
